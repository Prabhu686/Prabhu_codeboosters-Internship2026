```python
print('----------------------------------------------------------')
print('                Student Data Explorer')
print('----------------------------------------------------------')
print('Total Number of Students:',end=" ")
print(len(df))
print("Total Number of Columns:",end=" ")
print(df.shape[1])
print("Number of unique departments:",end=" ")
print(df['department'].nunique())
print("Number of Students per department:",end=" ")
print(df['department'].value_counts())
high_math=df['math_score'].max()
print(f'Highest math score:{high_math}')
low_math=df['math_score'].min()
print(f'Lowest math score:{low_math}')
avg_math=df['math_score'].mean()
print(f'Average math score:{avg_math}')
avg_prog=df['programming_score'].mean()
print(f'Avergae Programming Score:{avg_prog}')
avg_att=df['attendance_percentage'].mean()
print(f'Average Attendance Percentage:{avg_att}')
gender_avg=df.groupby('gender')['total_score'].mean().round(2)
print(f"Average Total Score grouped by gender:{gender_avg}")
df_sorted=df.sort_values(by='total_score',ascending=False)
print("Top 5 students by total score")
print(df_sorted[['name','department','total_score']].head(5))
low_att=df[df['attendance_percentage']<75]
print("Students with low attendance percentage:")
print(low_att[['name','department','attendance_percentage']])
```
